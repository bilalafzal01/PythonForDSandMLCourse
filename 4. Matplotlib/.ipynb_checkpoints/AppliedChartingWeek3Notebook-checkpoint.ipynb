## Histograms

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2, sharex=True)
axs = [ax1, ax2, ax3, ax4]

for n in range(0, len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    axs[n].hist(sample)
    axs[n].set_title('n={}'.format(sample_size))
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2, sharex=True)
axs = [ax1, ax2, ax3, ax4]

for n in range(0, len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    axs[n].hist(sample, bins=100)
    axs[n].set_title('n={}'.format(sample_size))
```

```python
plt.figure()
Y = np.random.normal(loc=0.0, scale=1.0, size=10000)
X = np.random.random(size=10000)
plt.scatter(X, Y)
```

```python
import matplotlib.gridspec as gridspec

plt.figure()
gspec = gridspec.GridSpec(3, 3)
top_histogram = plt.subplot(gspec[0,1:])
side_histogram = plt.subplot(gspec[1:, 0])
bottom_histogram = plt.subplot(gspec[1:, 1:])
```

```python
Y = np.random.normal(loc=0.0, scale=1.0, size=10000)
X = np.random.random(size=10000)
bottom_histogram.scatter(X, Y)
top_histogram.hist(X, bins=100)
s = side_histogram.hist(Y, bins=100, orientation='horizontal')
```

```python
top_histogram.clear()
top_histogram.hist(X, bins=100, density=True)
side_histogram.clear()
side_histogram.hist(Y, bins=100, orientation='horizontal', density=True)
# flip the side histogram's x axis
side_histogram.invert_xaxis()
```

```python
for ax in [top_histogram, bottom_histogram]:
    ax.set_xlim(0,1)
for ax in [side_histogram, bottom_histogram]:
    ax.set_ylim(-5,5)
```

## Box plots and whiskers

```python
import pandas as pd
normal_sample = np.random.normal(loc=0.0, scale=1.0, size=10000)
random_sample = np.random.random(size=10000)
gamme_sample = np.random.gamma(2, size=10000)

df = pd.DataFrame({'normal': normal_sample, 'random': random_sample, 'gamma': gamme_sample})
```

```python
df.describe()
```

```python
plt.figure()
_ = plt.boxplot(df['normal'], whis=[0,100])
```

```python
plt.clf()
_ = plt.boxplot([df['normal'], df['random'], df['gamma']], whis=[0,100])
```

```python
plt.figure()
_ = plt.hist(df['gamma'], bins=100)
```

```python
import mpl_toolkits.axes_grid1.inset_locator as mpl_il

plt.figure()
plt.boxplot([df['normal'], df['random'], df['gamma']], whis=[0,100])

ax2 = mpl_il.inset_axes(plt.gca(), width='60%', height='40%', loc=2)
ax2.hist(df['gamma'], bins=100)
ax2.margins(x=0.5)
```

```python
ax2.yaxis.tick_right()
```

```python
plt.figure()
_ = plt.boxplot([ df['normal'], df['random'], df['gamma'] ] )
```
